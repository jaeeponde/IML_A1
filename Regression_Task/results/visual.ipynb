```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
csv_file = "/Users/jaeeponde/IML_A1/IML_A1/Regression_Task/results/autopolypredictions.csv"  # Replace with the correct path to your CSV file
df = pd.read_csv(csv_file)

# Ensure your CSV has the columns 'Actual' and 'Predicted'
# If the column names are different, adjust them accordingly
actual_values = df['Actual'].values
predicted_values = df['Predicted'].values

# Plot Actual vs Predicted values
plt.figure(figsize=(8, 6))
plt.scatter(actual_values, predicted_values, color='blue', edgecolors='k', alpha=0.7, s=100)
plt.plot([min(actual_values), max(actual_values)], [min(actual_values), max(actual_values)], color='red', linestyle='--', linewidth=2)
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.grid(True)
plt.show()
```
